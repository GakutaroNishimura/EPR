```python
import ROOT
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./WaveData/scope_68.csv", names = ["time", "signal", "monitor", "sync"], skiprows=2)
```

```python
gr = ROOT.TGraph(len(df.time), np.array(df.time), np.array(df.signal))
c1 = ROOT.TCanvas("c1", "title", 400, 300)
gr.Draw("AP")
gr.SetTitle("Vsignal")
c1.Draw()
```

```python
#f = ROOT.TF1("func", "ROOT.Math.normal_cdf([0]*x, 1, 0)")
f = ROOT.TF1("func", "1-gaus(x, [0], [1], [2])*ROOT::Math::normal_cdf(x, 1, [1])", -5, 5)
f.SetParameters(1, 0, 1)
c = ROOT.TCanvas("c", "title", 400, 400)
f.Draw()
c.SetGridx()
c.Draw()
```
